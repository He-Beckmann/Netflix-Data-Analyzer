```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
import numpy
df = pd.read_csv('../ViewingActivity.csv')
df.columns = df.columns.str.replace(' ','_')
```

```python
df.shape
```

```python
seriesTitles = []
chapterTitles = []
episodeTitles = []
for title in df.Title:
    seriesTitle = title.split(':')[0]
    if len(title.split(':')) > 1:
        chapterTitle = title.split(':')[1]
    else:
        chapterTitle = 'NaN'  
    if len(title.split(':')) > 2:
        episodeTitle = title.split(':')[2]
    else:
        episodeTitle = 'NaN' 
    seriesTitles.append(seriesTitle)
    chapterTitles.append(chapterTitle)
    episodeTitles.append(episodeTitle)
    

df['Series_Title'] = seriesTitles
df['Chapter_Title'] = chapterTitles
df['Episode_Title'] = episodeTitles
```

```python
df.shape
```

```python
df.Duration = pd.to_timedelta(df.Duration)
# df.Duration = df.Duration.astype('timedelta64[s]')
df.Start_Time = pd.to_datetime(df.Start_Time)
```

```python
# Split Accounts
hendrik = df.query("Profile_Name == 'Hendrik'")
hendrik.shape
```

```python
# Remove Advertisment
hendrik = hendrik.query("Supplemental_Video_Type.isnull()", engine='python')
```

```python
hendrik.head(1)
```

```python
hendrik = hendrik.groupby(['Series_Title']).sum().sort_values(['Duration'],ascending=False)
hendrik.head(100)
```
